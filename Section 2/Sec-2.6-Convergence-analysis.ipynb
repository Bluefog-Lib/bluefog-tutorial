## 2.6 Convergence of Average Consensus

In Sec.2.2--2.5, we introduced the network topology, the combination matrix, and the consensus average algorithm. In this section, we will discuss the convergence of average consensus, and how network toplogy affect the convergence rate.

### 2.6.1 Riview of average consensus algorithm

Average consensus is a decentralized algorithm to let each agent reach the global average of each local variables. Given a network of $n$ computing agents, the combination matrix $W$, and the local variables $x_1^0, \cdots, x_n^0$ (each variable has dimension $d$), the average consensus algorithm is  

$$x_i^{k} = \sum_{j=1}^n w_{ij} x_j^{k-1} = w_{ii} x_i^{k-1} + \sum_{j\in \mathcal{N}(i)} w_{ij} x_j^{k-1}, \quad \forall\ i=1,\cdots, n.$$

Average consensus is a recursive algorithm. Each recursion only incurs neighborhood communication. Each recursion of average consensus is conducted in parallel. By repeating a series of partial averaging, it is guaranteed by theory that $x_i^{k} \to \bar{x}^0 = \frac{1}{n}\sum_{i=1}^n x_i^0$.

**Notation.** We introduce some notations to facilitate the analysis.

- We let ${X} = [x_1^T;\cdots;x_n^T]\in \mathbb{R}^{n\times d}$ to collect each local variable $x_i$ into a matrix, and $X^{(k)}$ be the $k$-th iterate of variable $X$;


- Given matrices $W\in \mathbb{R}^{n\times n}$ and $X\in \mathbb{R}^{n\times d}$, it holds that 

\begin{align}\label{submulti}
\|W X\|_F \le \|W\|_2 \|X\|_F.
\end{align}

To verify it, by letting $X_j$ be the $j$-th column of $X$, we have 

$$\|WX\|_F^2 = \sum_{j=1}^d \|X_j\|_2^2 \le \sum_{j=1}^d \|W\|_2^2 \|X_j\|_2^2=\|W\|_2^2\|X\|_F^2.$$ 


- Let $\lambda_i(A)$ be the $i$-th eigenvalue of the matrix $A$, and $\rho(A)$ be the spectrum radius of the matrix $A$.

### 2.6.2 Convergence analysis

**Reformulation of aveage consensus.** With notation $X^{(k)}$ and the combination matrix $W$, we can rewrite the avearge consensus algorithm as

$$X^{(k+1)} = W X^{(k)}$$


**Assumption on $W$.** Recall the assumption on the combination matrix $W$:

> **Assumption 1 (Doubly stochastic)** We assume $W$ is irreducible and non-periodic, and it satisfies $W \mathbf{1} = \mathbf{1}$ and $\mathbf{1}^T W = \mathbf{1}^T.$

**Properties of $W$.** If $W$ satisfies Assumption 1, it has the following properties:

> **Proposition 1** If matrix $W \in \mathbb{R}^{n\times n}$ satisfes Assumption 1, it has the following properties: 
>
> - The spectral radius of $W$ equals 1, i.e., $\rho(W) = \max\{|\lambda_1(A)|, |\lambda_2(A)|, \ldots,  |\lambda_n(A)|\} = 1$. Furthermore, there is one and only one eigenvalue with magnitude being $1$.
>
>
> - The spectral radius of $W-\frac{1}{n}\mathbf{1} \mathbf{1}^T$ is strictly less than $1$, i.e., $\rho(W-\frac{1}{n}\mathbf{1} \mathbf{1}^T) < 1$.

**Convergence of average consensus.** Give $X^{(0)} = [(x_1^{(0)})^T;\cdots;(x_n^{(0)})^T] \in \mathbb{R}^{n\times d}$, the goal of average consensus is to let $X^{(k)}$ converge to $\frac{1}{n}\mathbf{1}\mathbf{1}^TX^{(0)}$. As a result, we have

\begin{align*}
\|X^{(k+1)} - \frac{1}{n}\mathbf{1}\mathbf{1}^T X^{(0)}\|_F &= \|W^{k+1}X^{(0)} - \frac{1}{n}\mathbf{1}\mathbf{1}^TX^{(0)}\| \nonumber \\
&\le \|W^{k+1} - \frac{1}{n}\mathbf{1}\mathbf{1}^T\|\|X^{(0)}\|_F \nonumber \\
&= \|W - \frac{1}{n}\mathbf{1}\mathbf{1}^T\|^{k+1} \|X^{(0)}\|_F \nonumber \\
&= \rho^{k+1} \|X^{(0)}\|_F
\end{align*}

where $\rho = \rho(W - \frac{1}{n}\mathbf{1}\mathbf{1}^T) < 1$ because $W$ is doubly stochastic. The above derivation leads to the following theorem:

> **Theorem 1.** Under Assumption 1, $x_i^{(k)}$ generated by average consensus will converge to $\bar{x}^{(0)} = \frac{1}{n}\sum_{i=1}^n x_i^{(0)}$ at linear rate, i.e., 
>
> $$\|x_i^{(k)} - \bar{x}^{(0)}\| \le \rho^k \|X^{(0)}\|_F$$
>
> where $\rho = \rho(W - \frac{1}{n}\mathbf{1}\mathbf{1}^T) < 1$.
